In this notebook we invesitgate about the images of the ACDC dataset

```python
import os
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
import cv2
import matplotlib.pyplot as plt
import mpl_toolkits
```

```python
base_dir = r'C:\Users\alex1\Documents\Fundamentals_of_Data_Science\PFM\Datasets\ACDC Dataset\training_acdc\training'
```

We check thethe images for each disease

#### DCM: Dilated Cardiomyopathy

```python
data_1 = nib.load(os.path.join(base_dir, 'patient010','patient010_frame01.nii.gz'))
data_2 = nib.load(os.path.join(base_dir, 'patient010','patient010_frame01_gt.nii.gz'))
```

```python
data_1 = np.array(data_1.get_fdata())
data_2 = np.array(data_2.get_fdata())
```

```python
fig = plt.figure(figsize=(15,4))
fig.add_subplot(131)
plt.imshow(data_1[:,:,0])
plt.axis('off')
plt.title('Cine-MRI')
fig.add_subplot(132)
plt.imshow(data_2[:,:,0] )
plt.axis('off')
plt.title('Segmentation-Mask')
fig.add_subplot(133)
plt.imshow(np.multiply(data_1[:,:,0],data_2[:,:,0]))
plt.axis('off')
plt.title('MRI-Mask')
plt.show()
```

```python
fig = plt.figure(figsize=(6,2))
fig.add_subplot(131)
plt.imshow(np.multiply(data_1[:,:,8],np.equal(data_2[:,:,8],1)))
plt.axis('off')
plt.title('Right Ventricle')
fig.add_subplot(132)
plt.imshow(np.multiply(data_1[:,:,8],np.equal(data_2[:,:,8],2)))
plt.axis('off')
plt.title('Myocardium')
fig.add_subplot(133)
plt.imshow(np.multiply(data_1[:,:,8],np.equal(data_2[:,:,8],3)))
plt.axis('off')
plt.title('Left Ventricle')
plt.show()
```
